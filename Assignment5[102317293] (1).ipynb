```python
#QUESTION 1
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

summ = gfg.sum()

rsumm = gfg.sum(axis=1)

csumm = gfg.sum(axis=0)
print("Total sum of elements ", summ)
print("Row-wise sum:", rsumm)
print("Column-wise sum:", csumm)
```

```python
#QUESTION 2
import numpy as np

array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)

sorted_indices = np.argsort(array)

smallest_4 = np.sort(array)[:4]

largest_5 = np.sort(array)[-5:]

print("Sorted array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 smallest elements:", smallest_4)
print("5 largest elements:", largest_5)

```

```python
#QUESTION 3
import numpy as np
#a)

first_initial = 'V'
last_initial = 'K'

X = ord(first_initial) + ord(last_initial) #gets the asxii values


sales = np.array([X, X + 50, X + 100, X + 150, X + 200])

print("Sales data:", sales)


#b)

tax_rate = ((X % 5) + 5) / 100

tax_amount = sales * tax_rate
sales_after_tax = sales - tax_amount

print("Tax rate:", tax_rate)
print("Sales after tax:", sales_after_tax)

#c)
discounts = np.where(sales < (X + 100), 0.05, 0.10)
discounted_sales = sales_after_tax - (sales_after_tax * discounts)
print("Discounted Sales:", discounted_sales)

#d)

sales_matrix = np.vstack([discounted_sales] * 3)

week_multipliers = np.array([1.00, 1.02, 1.04]).reshape(3, 1)
adjusted_sales = sales_matrix * week_multipliers

print("Weekly adjusted sales:\n", adjusted_sales)
```

```python
#QUESTION 4

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)



y_square = x**2


y_sin = np.sin(x)


y_exp = np.exp(x)


y_log = np.log(np.abs(x) + 1)


plt.figure(figsize=(10, 6))


plt.plot(x, y_square, label=r'$y = x^2$', linestyle='-.', color='blue')


plt.plot(x, y_sin, label=r'$y = \sin(x)$', linestyle='-', color='red')


plt.plot(x, y_exp, label=r'$y = e^x$',linestyle=':', color='green')


plt.plot(x, y_log, label=r'$y = \log(|x| + 1)$',linestyle='--', color='purple')


plt.title("Plots of Various Functions")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid(True)

plt.show()
```
